```python
import pandas as pd
import sqlite3
```

```python
city_df = pd.read_csv('cities.csv', skipinitialspace=True)

conn = sqlite3.connect('era5_database.sqlite')
```

```python
df = pd.read_sql("SELECT * FROM main LIMIT 0", conn)

for _, row in city_df.iterrows():
    query = f"""SELECT * FROM main WHERE latitude = {row['Latitude']} AND longitude = {row['Longitude']}"""
    df = pd.concat([df, pd.read_sql(query, conn)], ignore_index=True)

df.head()
```

```python
df.to_sql('city_weather', conn, if_exists='replace', index=False)
df.to_csv('city_weather.csv', index=False)

conn.close()
```
