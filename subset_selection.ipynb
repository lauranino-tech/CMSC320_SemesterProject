# **Dataset Subset Selection**
The selected dataset contains over **13GB**. To mitigate this, we will be selecting specific subsets to analyze for the rest of the project.

**Chosen Subset:** We will be analyzing average monthly weather data for **30** locations around the world at noon from 2021 through the end of 2025. The list of cities (and stations) and their relative locations are stored in the *"cities.csv"* file. This file was creating through the personal selection of cities and stations from various continents as well as the online search (and rounding) of their coordinates.


**Step 1:** Import necessary libraries
- **pandas** for data handling
- **sqlite3** for SQL database management and querying

```python
import pandas as pd
import sqlite3
```

**Step 2:** Open *"cities.csv"* into a dataframe and connect to SQL database.

```python
city_df = pd.read_csv('cities.csv')

conn = sqlite3.connect('era5_database.sqlite')
```

**Step 3:** For each location in **city_df**, gather all its data from the dataset and combine it into a single dataframe

```python
df = pd.read_sql("SELECT * FROM main LIMIT 0", conn) # get column format

for _, row in city_df.iterrows(): # get all weather data for each location
    query = f"""SELECT * FROM main WHERE latitude = {row['latitude']} AND longitude = {row['longitude']}"""
    df = pd.concat([df, pd.read_sql(query, conn)], ignore_index=True)

df.shape
```

**Step 4:** Take the new dataset subset in *df* and add it as a table to the SQL database.

```python
df.to_sql('city_weather', conn, if_exists='replace', index=False)

conn.close()
```
