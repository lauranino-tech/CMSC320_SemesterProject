# **Data Cleaning**
Cleaning the selected data subset stored is necessary before exploration can begin. The dataset is stored in a SQL database, so the following must be **run first**: *sql_database_setup.py* and *subset_selection.ipynb*.


**Step 1:** Import necessary libraries
- **pandas** for data handling
- **sqlite3** for SQL database management and querying

```python
import pandas as pd
import sqlite3
```

**Step 2:** Connect to SQL database and store dataset in a dataframe.

```python
conn = sqlite3.connect('era5_database.sqlite')

weather_df = pd.read_sql("SELECT * FROM city_weather", conn)

weather_df.head()
```

**Step 3:** Keep important columns, discard the rest.

```python
weather_df = weather_df[['time', 'latitude', 'longitude', 'stl1', 'slt', 'sp', 'sd', 'u10', 'v10', 't2m', 'd2m', 'sf', 'e', 'ro', 'tp']]
```

**Step 4:** Display data types of **weather_df** columns.

```python
weather_df.dtypes
```

**Step 5:** Note that the 'time' column is stored as a string. We will convert it to a datetime object and only store the day, month, and year for convenience.

```python
weather_df['time'] = weather_df['time'].apply(
    lambda x : pd.to_datetime(x[0:10], format='%Y-%m-%d')
)

weather_df.head()
```

**Step 6:** Due to the way the data was recorded, each location and time has two entires. One for ground data, and the other for air (more specifically: percipitation related data). For each entry, we will combine it with its pair to remove missing data.

```python
weather_df = weather_df.groupby(['time', 'latitude', 'longitude']).max().reset_index()

weather_df.head()
```

**Step 7:** Rename columns for clarity.

```python
weather_df.columns = ['date', 'latitude', 'longitude', 'soil_temperature_level_1', 'soil_type', 'surface_pressure', 'snow_depth', 
    'u_wind_10m', 'v_wind_10m', 'temperature_2m', 'dewpoint_temperature_2m', 'snowfall', 'evaporation', 'runoff', 'total_percipitation']

weather_df.head()
```

**Step 8:** In order to add location name data to the dataset, we will use *cities.csv* which maps latitude and longitude to a city/location. First we need to import it into a dataframe.

```python
city_df = pd.read_csv('cities.csv')

city_df.head()
```

**Step 9:** Use **pandas** merge() function to combine the two dataframes on latitude and longitude.

```python
full_df = weather_df.merge(city_df, how='left', on=['longitude', 'latitude'])

full_df.head()
```

**Step 10:** From looking at the dtypes earlier, we note that soil_type is stored as a float64 despite containing int (whole number) data. soil_type is categorical data stored through ordinal encoding. While we are at it, we will reorder the column order to make things more ordered and grouped together visually.

```python
full_df['soil_type'] = full_df['soil_type'].astype('int64')

full_df = full_df[['location', 'continent', 'date', 'latitude', 'longitude', 'temperature_2m', 'surface_pressure', 'dewpoint_temperature_2m'
    , 'v_wind_10m', 'u_wind_10m', 'evaporation', 'runoff', 'total_percipitation', 'snowfall', 'snow_depth', 'soil_temperature_level_1', 'soil_type']]

full_df.head()
```

**Step 11:** Before saving this dataframe into a file, we will first check that everything looks alright.

```python
full_df.info()
```

**Step 12:** We notice that there are no null values, and that every column is stored as the desired dtype. Last thing to do is export the dataframe as a CSV file for ease of access.

```python
full_df.to_csv('weather.csv')
```
